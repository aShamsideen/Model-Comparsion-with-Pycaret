```python
import pycaret
import pandas as pd
from pycaret.classification import setup
from pycaret.classification import compare_models
```

```python
df = pd.read_csv('sonar.csv', header=None)
print(df.shape)
```

```python
n_cols = df.shape[1]
df.columns = [str(i) for i in range(n_cols)]
df.head()
```

```python
# setup the dataset
grid = setup(data=df, target=df.columns[-1], html=True, verbose=False)
best = compare_models()
print(best)
```
